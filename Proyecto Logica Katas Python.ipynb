1.-Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias
de cada letra en la cadena. Los espacios no deben ser considerados.


```python
def frecuencia_letras(texto):
    """Funcion para calcular la frecuencia ( numero de veces que aparece cada una de la letras del parametro texto)

    Args:
        texto (str): Cadena de texto como parámetro que queremos analizar
    return:
        Me devuleve el dicionario con las clave - letras y los valores - frecuencia /repeticion de la cadena de texto

    """
    frecuencia={} # Creo un diccionario vacío

    for letra in texto:
        if letra != " ": #si la letra es diferente de "vacio"=espacio
            if letra in frecuencia: #si la letra ya está en el diccionario frecuencia
                frecuencia [letra]= frecuencia [letra]+1 # Me sumas 1 a la fecuencia por que ya está
            else: #si no está la letra en el diciconario frecuencia
                frecuencia [letra]=1 # me considera frecuencia 1
    return frecuencia
```

3.-Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros. La función debe 
devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo

```python
def palabras_objetivo (lista, objetivo):
    """Funcion que buscará la palabra objetivo dentro de la lista
    Args:
    lista (list): lista de palabras aleatoria
    objetivo(str): palabra que deberemos buscar dentro de la lista de palabras
    return:
    Lista de palabras que contienen la palabra objetivo
    """
    resultado=[] # Aqui iremos metiendo las palabras que cumplan con la condicion

    for palabra in lista: # bucle para pasar por todas las palabras de la lista
        if objetivo in palabra: #Si el objetivo está contenido en la palabra
            resultado.append(palabra) #añadimos las palabras que cumplan el objetivo a la lista resultado
    return resultado
```

5.-Ecribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por 
defecto es 5. La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual 
que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función debe devolver 
una tupla que contenga la media y el estado

```python
def calculo_medias (notas, aprobado=5):
    suma = 0 #debemos decirle un punto desde donde partir para ir sumando las notas
    for nota in notas: #hacemos bucle for para ir recorriendo cada nota y hacer lo siguiente
        suma= suma + nota #vamos sumando las notas que vamos enocntrando con el bucle for
        media= suma/len(notas) #calculamos la media de esas notas (usamos len para saber la cantidad de valores que hay dentro de la lista notas)
    if media >=aprobado: #ponemos los condicionales
        estado = "aprobado"
       
    else:
        estado = "suspenso"
       
    return (round(media,2), estado) #para acotar los decimales ponemos la funcion round 
```

```python
notas=[3,5,7,1,6,8,10] #damos los valores al parametro notas
#invocamos a la funcion
medias=calculo_medias(notas,aprobado=5)
print (medias)
```

6.-Escribe una función que calcule el factorial de un número de manera recursiva


```python
def factorial (numero):
    """_Esta funcion calculara el factorial de un numero de forma recursiva
    args:
    numero (int): indicamos el nuemro a partir del cual debera iniciarse la factorizacion
    return:
    numero*(nuemro-1)
    """

    if numero == 1: # tope del factorial , si es 1 me debe devolver 1
        return 1
    else:
         return numero*factorial(numero-1) #si no es 1 , me devolvera el numero inicial por el numero anterior, asi sucesivamente 
    
 
```

mismas.
Caso de uso:

2. Hacer crecer el tronco del árbol una unidad.
3. Añadir una nueva rama al árbol.
4. Hacer crecer todas las ramas del árbol una unidad.
5. Añadir dos nuevas ramas al árbol.
6. Retirar la rama situada en la posición 2.
7. Obtener información sobre el árbol

```python
class Arbol:  #creamos la clase arbol , con la primera letra en mayusculas y seguido de dos puntos . Nohay parentesis
    def __init__(self,tronco,ramas): #creo el metodo constructor con parametros, ramas en el ejemplo será una lista
        self.tronco=tronco
        self.ramas=ramas
    def crecer_tronco(self): #creamos metodo para hacer aumentar la longitud 1 del tronco
        self.tronco += 1
    def nueva_rama(self): #creamos metodo para hacer crear una nueva rama
        self.ramas.append(1) #append es una funcion y por tanto el parametro va entre parentesis
    def crecer_ramas(self):#creamos metodo para hacer crecer todas las ramas
        for i in range(len(self.ramas)): #tiene que pasar por cada una de las ramas de la lista , por tanto debemos usar un bucle for in
            self.ramas[i]+=1
    def quitar_rama (self, posicion): #creamos metodo para quitar una rama de una posicion especifica.Usaremos la funcion pop
        self.ramas.pop(posicion)
    def informacion_arbol (self):
        print(f"la longitud del tronco es:{self.tronco}")
        print(f"el numero de ramas es:{len(self.ramas)}")
        print(f"la longitud de las ramas es:{self.ramas}")
```

```python
class UsuarioBanco: #creamos la clase UsuarioBanco , con la primera letra en mayusculas y seguido de dos puntos . No hay parentesis
    def __init__(self, nombre,saldo,cuenta_corriente): #definimos el metido constructor con 3 parametros nombre, saldo, cuenta_corriente
        self.nombre=nombre
        self.saldo=saldo
        self.cuenta_corriente=cuenta_corriente
    def agregar_dinero (self, cantidad): #definimos un metodo para agregar dinero, usamos como parametro la cantidad
        self.saldo+=cantidad
    def retirar_dinero (self,cantidad): #definimos metodo para retirar dinero , usamos como parametro la cantidad, y establecemos una condicion
        if cantidad>self.saldo: 
            print("no hay saldo suficiente")
        else:
            self.saldo+=cantidad
    def transferencia_dinero (self, otro_usuario, cantidad): #creamos metodo para transferir dinero a otro usuario, usamos como parametro , el otro usurio y la cantidad
        if cantidad>otro_usuario.saldo:
            print("El saldo del otro usurio no es suficiente")
        else:
            self.saldo+=cantidad  #sumamos la cantidad al saldo del usurio
            otro_usuario.saldo-=cantidad #restamos la cantidad en el saldo del otro usuario
        
     
```

```python
def contar_palabras(texto): #definimos la funcion contar palabras
    palabras=texto.split() #separmos las palabras del texto con la funcion split
    diccionario={} #creamos un diccionario vacio para ir almacenando las palabras que sacamos del texto

    for palabra in palabras: #hacemos bucle para ir pasando por cada palabra del texto
        if palabra in diccionario:
            diccionario[palabra]+=1 #le indocamos que si ya hay una palabra en el diccionario la siguiente que añada debe sumarle +1 (para que ya sean 2 y asi sucesivamnete) 
        else:
            diccionario[palabra]=1 #sino será la primera palabra por tanto deberá considerar 1
    return diccionario

```
